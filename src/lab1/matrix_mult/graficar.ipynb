```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os, os.path
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
data = pd.read_csv('values.csv')

x = data["Size"]
y1 = data["Tkrnl"]
y2 = data["BWkrnl"] / 1000

plt.plot(x, y1, color='red',marker='o',markerfacecolor='blue')
plt.xlabel("Tamaño de las matrices (NxN)")
plt.ylabel("Tiempo de ejecución del kernel (s)")
plt.show()
plt.plot(x, y2, color='green', marker='o', markerfacecolor='blue')
plt.xlabel("Tamaño de las matrices (NxN)")
plt.ylabel("Eficiencia del kernel (GFLOPS)")
plt.show()
```
