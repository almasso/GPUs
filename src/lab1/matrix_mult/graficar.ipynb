```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os, os.path
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
iterations = 64
data = pd.read_csv('values.csv')

x = data["Size"]
y1 = data["Tkrnl"]
y2 = data["BWkrnl"]
titles = ["Un hilo", "Multihilo", "Multihilo eficiente"]

for i in range(2):
    plt.plot(x[iterations * i:iterations * (i+1)], y1[iterations * i:iterations * (i+1)],marker='o',label=titles[i + 1])
plt.xlabel("Tamao de las matrices (NxN)")
plt.ylabel("Tiempo de ejecucin del kernel (s)")
plt.legend()
plt.show()
for i in range(2):
    plt.plot(x[iterations * i:iterations * (i+1)], y2[iterations * i:iterations * (i+1)], marker='o',label=titles[i + 1])
plt.xlabel("Tamao de las matrices (NxN)")
plt.ylabel("Eficiencia del kernel (GFLOPS)")
plt.legend()
plt.show()
```
